```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
```

```python
tickers = ['GOOG','AMZN','AAPL']
histories = {}
recommends = {}
for ticker in tickers:
    df_history = pd.read_csv('data/original/%s_history.csv'%ticker)
    df_history['Date'] = pd.to_datetime(df_history['Date'])
    histories[ticker] = df_history
    df_recommends = pd.read_csv('data/original/%s_recommends.csv'%ticker)
    df_recommends['Date'] = pd.to_datetime(df_recommends['Date'])
    recommends[ticker] = df_recommends
histories['GOOG'].head()
```

```python
recommends['AMZN'].head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 5))
for ticker in tickers:
    stock_df = histories[ticker]
    ax.plot(stock_df['Date'],stock_df['Open'], label=ticker)
ax.legend(loc='upper left')
ax.set_xlabel('Open')
ax.set_ylabel('Date')
fig.suptitle("Open pices over past decade")
fig.savefig("outputs/opens.png", dpi=200)
```
